```python
import tkinter as tk
from tkinter import messagebox

def ingresar_segunda_ventana():
    ventana.withdraw()  # Ocultar la ventana actual
    segunda_ventana.deiconify()  # Mostrar la segunda ventana

def regresar_primera_ventana():
    segunda_ventana.withdraw()  # Ocultar la segunda ventana
    ventana.deiconify()  # Mostrar la primera ventana

def procesar_distrito_origen():
    global distrito_origen
    distrito_origen = entry_distrito_origen.get()
    segunda_ventana.withdraw()  # Ocultar la segunda ventana
    tercera_ventana.deiconify()  # Mostrar la tercera ventana

def regresar_segunda_ventana():
    tercera_ventana.withdraw()  # Ocultar la tercera ventana
    segunda_ventana.deiconify()  # Mostrar la segunda ventana

def procesar_distrito_final():
    global distrito_final
    distrito_final = entry_distrito_final.get()
    tercera_ventana.withdraw()  # Ocultar la tercera ventana
    cuarta_ventana.deiconify()  # Mostrar la cuarta ventana

def mostrar_datos_aplicativo():
    ventana_cinco = tk.Toplevel(cuarta_ventana)
    ventana_cinco.title("Ventana Cinco")
    ventana_cinco.geometry(f"{ancho_ventana}x{alto_ventana}")
    
    label_datos_aplicativo = tk.Label(ventana_cinco, text="DATOS DEL APLICATIVO SELECCIONADO:")
    label_datos_aplicativo.pack()

    label_punto_partida = tk.Label(ventana_cinco, text="Punto de partida:")
    label_punto_partida.pack()
    label_punto_partida_valor = tk.Label(ventana_cinco, text="N/D")
    label_punto_partida_valor.pack()

    label_destino_final = tk.Label(ventana_cinco, text="Destino final:")
    label_destino_final.pack()
    label_destino_final_valor = tk.Label(ventana_cinco, text="N/D")
    label_destino_final_valor.pack()

    label_aplicativo = tk.Label(ventana_cinco, text="Aplicativo:")
    label_aplicativo.pack()
    label_aplicativo_valor = tk.Label(ventana_cinco, text="N/D")
    label_aplicativo_valor.pack()

    label_precio = tk.Label(ventana_cinco, text="Precio:")
    label_precio.pack()
    label_precio_valor = tk.Label(ventana_cinco, text="N/D")
    label_precio_valor.pack()

    label_confirmar = tk.Label(ventana_cinco, text="CONFIRMAR?")
    label_confirmar.pack()

    frame_botones_confirmar = tk.Frame(ventana_cinco)
    frame_botones_confirmar.pack()

    btn_si = tk.Button(frame_botones_confirmar, text="Si", command=mostrar_ventana_seis)
    btn_si.pack(side="left", padx=5, pady=5)

    btn_no = tk.Button(frame_botones_confirmar, text="No", command=ventana_cinco.destroy)
    btn_no.pack(side="left", padx=5, pady=5)

def mostrar_ventana_seis():
    ventana_seis = tk.Toplevel(cuarta_ventana)
    ventana_seis.title("Ventana Seis")
    ventana_seis.geometry(f"{ancho_ventana}x{alto_ventana}")

    label_gracias = tk.Label(ventana_seis, text="MUCHAS GRACIAS POR LA RESERVA")
    label_gracias.pack()

    label_contactaremos = tk.Label(ventana_seis, text="A continuación, contactaremos con un socio conductor de la aplicación que usted seleccionó, para dar inicio a su viaje.")
    label_contactaremos.pack()

    label_llamada = tk.Label(ventana_seis, text="En breve recibirá una llamada del conductor....")
    label_llamada.pack()

    btn_aceptar = tk.Button(ventana_seis, text="Aceptar", command=cerrar_todas_ventanas)
    btn_aceptar.pack()

def cerrar_todas_ventanas():
    ventana_seis.quit()
    ventana.quit()

# Tamaño fijo para todas las ventanas
ancho_ventana = 400
alto_ventana = 250

# Crear la ventana principal
ventana = tk.Tk()
ventana.title("EasyRoutE - Comparador de precios de taxi")
ventana.geometry(f"{ancho_ventana}x{alto_ventana}")

# Crear los elementos de la interfaz de la primera ventana
label_bienvenida = tk.Label(ventana, text="Bienvenidos a la aplicación EasyRoutE", padx=10, pady=10)
label_bienvenida.pack()

label_descripcion1 = tk.Label(ventana, text="Con esta aplicación, usted", padx=10, pady=1)
label_descripcion1.pack()

label_descripcion2 = tk.Label(ventana, text="podrá seleccionar su viaje en taxi ideal", padx=30)
label_descripcion2.pack()

btn_ingresar = tk.Button(ventana, text="Ingresar", command=ingresar_segunda_ventana, padx=10, pady=10)
btn_ingresar.pack()

# Crear la segunda ventana (oculta por defecto)
segunda_ventana = tk.Toplevel(ventana)
segunda_ventana.title("Segunda Ventana")
segunda_ventana.geometry(f"{ancho_ventana}x{alto_ventana}")
segunda_ventana.withdraw()  # Ocultar la segunda ventana inicialmente

# Crear los elementos de la segunda ventana
label_distrito_origen = tk.Label(segunda_ventana, text="El distrito origen es:")
label_distrito_origen.pack()

entry_distrito_origen = tk.Entry(segunda_ventana)
entry_distrito_origen.pack()

frame_botones_segunda = tk.Frame(segunda_ventana)
frame_botones_segunda.pack()

btn_regresar_segunda = tk.Button(frame_botones_segunda, text="Regresar", command=regresar_primera_ventana)
btn_regresar_segunda.pack(side="left", padx=5, pady=5)

btn_aceptar_segunda = tk.Button(frame_botones_segunda, text="Aceptar", command=procesar_distrito_origen)
btn_aceptar_segunda.pack(side="left", padx=5, pady=5)

# Crear la tercera ventana (oculta por defecto)
tercera_ventana = tk.Toplevel(ventana)
tercera_ventana.title("Tercera Ventana")
tercera_ventana.geometry(f"{ancho_ventana}x{alto_ventana}")
tercera_ventana.withdraw()  # Ocultar la tercera ventana inicialmente

# Crear los elementos de la tercera ventana
label_distrito_final = tk.Label(tercera_ventana, text="El distrito final es:")
label_distrito_final.pack()

entry_distrito_final = tk.Entry(tercera_ventana)
entry_distrito_final.pack()

frame_botones_tercera = tk.Frame(tercera_ventana)
frame_botones_tercera.pack()

btn_regresar_tercera = tk.Button(frame_botones_tercera, text="Regresar", command=regresar_segunda_ventana)
btn_regresar_tercera.pack(side="left", padx=5, pady=5)

btn_aceptar_tercera = tk.Button(frame_botones_tercera, text="Aceptar", command=procesar_distrito_final)
btn_aceptar_tercera.pack(side="left", padx=5, pady=5)

# Crear la cuarta ventana (oculta por defecto)
cuarta_ventana = tk.Toplevel(ventana)
cuarta_ventana.title("Cuarta Ventana")
cuarta_ventana.geometry(f"{ancho_ventana}x{alto_ventana}")
cuarta_ventana.withdraw()  # Ocultar la cuarta ventana inicialmente

# Crear los elementos de la cuarta ventana
label_tarifas = tk.Label(cuarta_ventana, text="Lista recomendada de tarifas que ofrecen los aplicativos de taxis en Lima:")
label_tarifas.pack()

# Añadir el recuadro de muestra (puedes personalizarlo según tus necesidades)
recuadro_muestra = tk.Text(cuarta_ventana, width=50, height=10)
recuadro_muestra.pack()

frame_botones_cuarta = tk.Frame(cuarta_ventana)
frame_botones_cuarta.pack()

btn_regresar_cuarta = tk.Button(frame_botones_cuarta, text="Regresar")
btn_regresar_cuarta.pack(side="left", padx=5, pady=5)

btn_seleccionar = tk.Button(frame_botones_cuarta, text="Seleccionar", command=mostrar_datos_aplicativo)
btn_seleccionar.pack(side="left", padx=5, pady=5)

btn_siguiente = tk.Button(frame_botones_cuarta, text="Siguiente")
btn_siguiente.pack(side="left", padx=5, pady=5)

# Variables para almacenar los distritos origen y destino
distrito_origen = ""
distrito_final = ""

# Iniciar el bucle principal de la ventana
ventana.mainloop()

```
